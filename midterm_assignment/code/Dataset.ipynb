````python
#Dataset1
n = 100
x = 3*(np.random.rand(n,2)-0.5)
radius = x[:,0]**2+x[:,1]**2
y = (radius > 0.7 + 0.1 + np.random.randn(n)) & (radius < 2.2 + 0.1 + np.random.randn(n))
y = 2 * y - 1
````

````python
plt.scatter(x[:,0],x[:,1],color=np.where(y>0,"r","b"))
````

Dataset IV
```
n = 200;
x = 3 * (rand(n, 4) - 0.5);

```

````python
#Dataset4
n = 200
x = 3 * (np.random.rand(n,4) - 0.5)
y = (2 * x[:,0] - 1 * x[:,1] + 0.5 + 0.5 * np.random.rand(n)) > 0
y = 2 * y - 1
````

````python
print(x.shape,y.shape)

fig,axs = plt.subplots(4,4,figsize=(10,10),sharex=True,sharey=True)
for i in range(4):
    for j in range(4):
        axs[i,j].scatter(x[:,i],x[:,j],color=np.where(y>0,"r","b"))
plt.tight_layout()
plt.show()
````

Dataset V
```
n = 200;
x = 3 * (rand(n, 4) - 0.5);
W = [2, -1, 0.5;
     -3, 2, 1;
      1, 2, 3];
[maxlogit, y] =max( [x(:, 1:2), ones(n, 1)] * W’ + 0.5 * randn(n, 3), [], 2);
```

````python
#Dataset5
n = 200
x = 3 * (np.random.rand(n,4) - 0.5)
W = [[ 2, -1,0.5],
     [-3,  2,  1],
     [ 1,  2,  3]]
W = np.array(W)

tmp = np.hstack([x[:,:2],np.ones((n,1))])@W.T + 0.5 * np.random.randn(n,3)
maxlogit = np.amax(tmp,axis=1)
y = np.argmax(tmp,axis=1)
````

````python
print(x.shape,y.shape)

fig,axs = plt.subplots(4,4,figsize=(10,10),sharex=True,sharey=True)
colors = [["r","g","b"][i] for i in y]
for i in range(4):
    for j in range(4):
        axs[i,j].scatter(x[:,i],x[:,j],color=colors)
plt.tight_layout()
plt.show()
````
